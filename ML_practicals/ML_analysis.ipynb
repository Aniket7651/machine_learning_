```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x**4
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.scatter(dataf.A, dataf.D, c='red')
plt.show()
```

```python
fg, ax = plt.subplots(nrows=1, ncols=2)
for a in ax:
      a.plot(x, y, 'r+')
fg.tight_layout()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.6, 0.3, 0.2])
axes3 = fig.add_axes([0.2, 0.3, 0.3, 0.2])
x = np.linspace(0, 5)
y1 = x**4
y2 = x**(6+y1**0.5)
y3 = x**8
axes1.plot(x, y1, 'r')
axes2.plot(x, y2, 'g')
axes3.plot(x, y3, 'b')
plt.show()
```
